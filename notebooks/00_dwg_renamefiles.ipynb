```python
from pathlib import Path
from tqdm import tqdm

# Define the directory containing the files
directory = Path('/app/data/raw/v0.40/')

# Define the replacement patterns
replacements = {
    'sentinel_2': 'S2L2A',
    'sentinel_1': 'S1RTC',
    'dem': 'DEM'
}

# Get list of files to rename (for accurate progress bar)
files_to_rename = [
    f for f in directory.iterdir() 
    if f.is_file() and any(pattern in f.name for pattern in replacements.keys())
]

# Iterate through all files with progress bar
for file_path in tqdm(files_to_rename, desc="Renaming files"):
    # Create new filename by applying all replacements
    new_name = file_path.name
    for old_pattern, new_pattern in replacements.items():
        new_name = new_name.replace(old_pattern, new_pattern)
    
    new_path = file_path.parent / new_name
    
    # Rename the file
    file_path.rename(new_path)
```
